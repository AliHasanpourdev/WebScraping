```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.options import Options
import csv
import time
```

```python
options = Options()
options.add_argument("--headless")
driver = webdriver.Firefox(options=options)
url = "https://enamad.ir/DomainListForMIMT/Index/"
```

```python
nexpath = "/html[1]/body[1]/div[1]/div[2]/div[2]/div[1]/div[1]/section[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/ul[1]/div[1]/div[1]/div[2]/div[2]/div[3]/div[1]/ul[1]/li[12]/a[1]"
headers = ["ردیف", "آدرس دامنه", "عنوان کسب و کار", "استان", "شهر", "تاریخ اعطا", "تاریخ انقضا"]
with open("URLs.csv", "+a") as file :
    writer = csv.writer(file)
    writer.writerow(headers)
for k in range(2, 6668) :
    driver.get(url+str(k))
    for i in range(30) :
        rows = []
        for j in range(8) :
            if j==5 :
                continue
            elif j==1 :
                xpath = f"/html[1]/body[1]/div[1]/div[2]/div[2]/div[1]/div[1]/section[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/ul[1]/div[1]/div[1]/div[2]/div[2]/div[2]/div[{i+1}]/div[{j+1}]"
                rows.append("https://" + driver.find_element(By.XPATH, xpath).text)
            else :
                xpath = f"/html[1]/body[1]/div[1]/div[2]/div[2]/div[1]/div[1]/section[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/ul[1]/div[1]/div[1]/div[2]/div[2]/div[2]/div[{i+1}]/div[{j+1}]"
                rows.append(driver.find_element(By.XPATH, xpath).text)
        with open("URLs.csv", "+a") as file :
            writer = csv.writer(file)
            writer.writerow(rows)
    time.sleep(10)
```
